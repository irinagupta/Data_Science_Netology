# HOGDescriptor

https://docs.opencv.org/2.4/modules/gpu/doc/object_detection.html

```python
import cv2
import numpy as np
print(cv2.__version__)
```

## Загружаем исходные данные

```python
train = np.loadtxt('data/digit/train.csv', delimiter=',', skiprows=1)
# сохраняем разметку в отдельную переменную
train_label = train[:, 0]
# приводим размерность к удобному для обаботки виду
train_img = np.resize(train[:, 1:], (train.shape[0], 28, 28))
```

## Задаем параметры для вычисления дескриптора

```python
winSize = (28,28)
blockSize = (14,14)
blockStride = (14,14)
cellSize = (7,7)
nbins = 9
derivAperture = 1
winSigma = 4.
histogramNormType = 0
L2HysThreshold = 2.0000000000000001e-01
gammaCorrection = 0
nlevels = 64

hog = cv2.HOGDescriptor(winSize,
                        blockSize,
                        blockStride,
                        cellSize,
                        nbins,
                        derivAperture,
                        winSigma,
                        histogramNormType,
                        L2HysThreshold,
                        gammaCorrection,
                        nlevels)
```

## Вычисляем дескриптор для изображения

```python
descriptor = hog.compute(np.uint8(train_img[0]))
print('Descriptor size: %d' % len(descriptor))
```

```python
from matplotlib import pyplot as plt

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(range(len(descriptor)), descriptor.reshape(-1))
plt.show()
```

```python
def show(img_number):
    descriptor = hog.compute(np.uint8(train_img[img_number]))
    fig = plt.figure(figsize=(20,10))
    subplot = fig.add_subplot(1, 2, 1)
    plt.imshow(train_img[img_number], cmap='gray')
    subplot = fig.add_subplot(1, 2, 2)
    plt.bar(range(len(descriptor)), descriptor.reshape(-1))
    plt.show()
    
    return descriptor
```

```python
show(0)
```

```python
desc1 = show(1)
```

```python
show(2)
```

```python
desc5 = show(3)
```

```python
desc2 = show(4)
```

```python
desc3 = show(5)
```

```python
desc4 = show(6)
```

```python
show(8)
```

```python
descs = [desc1, desc2, desc3, desc4, desc5]
```

```python
target = desc1
for i, d in enumerate(descs):
    print(i, ((target - d) ** 2).sum())
```

```python
images = []
for i in range(100):
    descriptor = hog.compute(np.uint8(train_img[i]))
    if ((target - descriptor) ** 2).sum() < 1.5:
        images.append(train_img[i])
```
