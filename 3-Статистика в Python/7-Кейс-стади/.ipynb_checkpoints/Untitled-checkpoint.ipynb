```python
import pandas as pd
from scipy import stats as st
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('video_games_sales.csv')
df
```

```python
df.info()
```

```python
df = df[['Genre', 'Platform', 'Critic_Score']].dropna()
df
```

Ответьте на следующие вопросы:

1) Как критики относятся к спортивным играм?

Для каждого вопроса:

сформулируйте нулевую и альтернативную гипотезы;

выберите пороговый уровень статистической значимости;

опишите полученные результаты статистического теста.

#### H0: критики довольны спортивными играми и средняя оценка выше 7 

#### H1: критики  не довольны спортивными играми и средняя оценка ниже 7

#### За alpha везде будем принимать 0.05

```python
alpha = 0.05
```

```python
result = st.ttest_1samp(df[df['Genre'] == 'Sports']['Critic_Score'], 7)
print(result)

if (result.statistic < 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, критики не довольны спортивными играми')
else:
    print('Не отвергаем нулевую гипотезу, критики довольны спортивными играми')
```

2) Критикам нравятся больше игры на PC или на PS4?

сформулируйте нулевую и альтернативную гипотезы;

выберите пороговый уровень статистической значимости;

опишите полученные результаты статистического теста.

#### H0: Критикам нравятся больше игры на PC, средняя оценка за игры н PC, выше, чем за игры на PS4

#### H1: Критикам нравятся больше игры не на PC, а на PS4

```python
result = st.ttest_ind(df[df['Platform'] == 'PC']['Critic_Score'], df[df['Platform'] == 'PS4']['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha) & (result.statistic < 0):
    print('Отвергаем нулевую гипотезу, Критикам нравятся больше игры не на PC, а на PS4')
else:
    print('Не отвергаем нулевую гипотезу')
```

3) Критикам больше нравятся стрелялки или стратегии?

сформулируйте нулевую и альтернативную гипотезы;

выберите пороговый уровень статистической значимости;

опишите полученные результаты статистического теста.

#### H0: Критикам нравятся больше стрелялки, средняя оценка за стрелялки, выше, чем за стратегии

#### H1: Критикам нравятся больше не стрелялки, а стратегии

```python
result = st.ttest_ind(df[df['Genre'] == 'Shooter']['Critic_Score'], df[df['Genre'] == 'Strategy']['Critic_Score'], equal_var=False)
print(result)
if (result.pvalue < alpha) & (result.statistic < 0):
    print('Отвергаем нулевую гипотезу, Критикам нравятся больше не стрелялки, а стратегии')
else:
    print('Не отвергаем нулевую гипотезу')
```

Задание 2

Реализуйте базовую модель логистической регрессии для классификации текстовых сообщений (используемые данные здесь) по признаку спама. 

```python
df_spam = pd.read_csv('spam.csv')
df_spam
```

Для этого:

1) Привидите весь текст к нижнему регистру;

```python
df_spam['Message'] = df_spam['Message'].str.lower()
df_spam
```

2) Удалите мусорные символы;

```python
df_spam['Message'] = df_spam['Message'].replace('[\W_]+', ' ', regex=True)
df_spam
```

3) Удалите стоп-слова;

```python
df_spam['Message'] = df_spam['Message'].str.split()
stopwords_set = set(stopwords.words('english'))
df_spam['Message'] = df_spam['Message'].apply(lambda x: [word for word in x if word not in stopwords_set])
df_spam
```

4) Привидите все слова к нормальной форме;

```python
wordnet_lemmatizer = WordNetLemmatizer()
df_spam['Message'] = df_spam['Message'].apply(lambda x: [wordnet_lemmatizer.lemmatize(word) for word in x])
df_spam
```

5) Преобразуйте все сообщения в вектора TF-IDF. Вам поможет следующий код:

Можете поэкспериментировать с параметрами TfidfVectorizer;

```python
# списки  вернем обратно в строку
df_spam['Message'] = df_spam['Message'].str.join(" ")
df_spam
```

```python
# Преобразуем все сообщения в вектора TF-IDF
tfidf = TfidfVectorizer(min_df=70)
tfidf_matrix = tfidf.fit_transform(df_spam['Message'])
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
```

6) Разделите данные на тестовые и тренировочные в соотношении 30/70, укажите random_state=42. Используйте train_test_split;

```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, df_spam['Category'], test_size=0.3, random_state=42)
```

7) Постройте модель логистической регрессии, укажите random_state=42, оцените ее точность на тестовых данных;

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)
lda.predict(X_test)
predict_all = pd.DataFrame(lda.predict(tfidf_matrix))
predict_all
```

8) Опишите результаты при помощи confusion_matrix;

```python
confusion_matrix(df_spam['Category'], predict_all)
```

```python
accuracy_score(df_spam['Category'], predict_all)
```

#### Результаты модели - 96%, довольно точно 

9) Постройте датафрейм, который будет содержать все исходные тексты сообщений, классифицированные неправильно (с указанием фактического и предсказанного).
