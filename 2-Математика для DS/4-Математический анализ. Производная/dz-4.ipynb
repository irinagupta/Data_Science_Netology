Уровень 0
посчитайте значение 1-ой и 2-ой производной фунции
f(x) = x^5 + 4sin(2x) + cos(3x+3) в точке x = 1

```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 5 + 4 * np.sin(2 * x) + np.cos(3 * x + 3)

x = 1
f_x = f(x)
f_1_x = derivative(f, x, dx=1e-6, n=1)
f_2_x = derivative(f, x, dx=1e-6, n=2)

print(f"Значение функции в точке x = {x} равно {f_x}.")
print(f"Значение 1-ой производной в точке x = {x} равно {f_1_x}.")
print(f"Значение 2-ой производной в точке x = {x} равно {f_2_x}.")
```

Уровень 1**
Постройте график первой и второй производной функции
(sin(2x+1))^5
на отрезке [-5; 5]

```python
def f(x):
    return (np.sin(2 * x + 1)) ** 5

x = np.linspace(-5, 5)
f_x = f(x)
f_1_x = [derivative(f, xx, dx=1e-6, n=1) for xx in x]
f_2_x = [derivative(f, xx, dx=1e-6, n=2) for xx in x]

plt.plot(x, f_x, label="График функии (sin(2x+1))^5")
plt.plot(x, f_1_x, label="График первой производной")
plt.plot(x, f_2_x, label="График второй производной")
plt.title("График функции и ее 1-ой и 2-ой производных")
plt.legend()
plt.show()
```
