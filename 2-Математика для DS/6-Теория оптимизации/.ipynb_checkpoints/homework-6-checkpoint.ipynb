### Задание

При фиксированном seed=42 поразбирайтесь и поэкспериментируйте с параметрами алгоритма дифференциальной эволюции:
strategy, popsize, tol, mutation, recombination и updating.

Постройте графики количества итераций оптимизации функции ackley от значения параметра.



```python
import numpy as np
from scipy.optimize import differential_evolution
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]
result = differential_evolution(ackley, bounds, seed=42)
result
```

```python
strategys = ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp', 
             'currenttobest1exp', 'best2exp', 'rand2exp', 'randtobest1bin', 'currenttobest1bin', 
             'best2bin',  'rand2bin', 'rand1bin']
result_strategy = []
for strategy in strategys:
    result_strategy.append(differential_evolution(ackley, bounds, strategy=strategy, seed=42))

data_strategy = pd.DataFrame(result_strategy).join(pd.DataFrame(strategys))
data_strategy.columns=['x', 'fun', 'nfev', 'nit', 'message', 'success', 'strategy']
data_strategy
```

```python
plt.barh(data_strategy['strategy'], data_strategy['nit'])
```

```python
popsizes = np.arange(0,500,50)
result_popsize = []
for popsize in popsizes:
    result_popsize.append(differential_evolution(ackley, bounds, popsize=popsize, seed=42))

data_popsize = pd.DataFrame(result_popsize).join(pd.DataFrame(popsizes))
data_popsize.columns=['x', 'fun', 'nfev', 'nit', 'message', 'success', 'jac', 'popsize']
data_popsize
```

```python
plt.bar(data_popsize['popsize'], data_popsize['nit'])
```

```python
tols = np.linspace(0,1,100)
result_tol = []
for tol in tols:
    result_tol.append(differential_evolution(ackley, bounds, tol=tol, seed=42))

data_tol = pd.DataFrame(result_tol).join(pd.DataFrame(tols))
data_tol.columns=['x', 'fun', 'nfev', 'nit', 'message', 'success', 'jac', 'tol']
data_tol
```

```python
plt.plot(data_tol['tol'], data_tol['nit'])
```

```python
mutations = np.linspace(0,2,100)
result_mutation = []
for mutation in mutations[0:99]:
    result_mutation.append(differential_evolution(ackley, bounds, mutation=mutation, seed=42))
data_mutation = pd.DataFrame(result_mutation).join(pd.DataFrame(mutations))
data_mutation.columns=['x', 'fun', 'nfev', 'nit', 'message', 'success', 'jac', 'mutation']
data_mutation
```

```python
plt.plot(data_mutation['mutation'], data_mutation['nit'])
```

```python
recombinations = np.linspace(0,1,100)
result_recombination = []
for recombination in recombinations:
    result_recombination.append(differential_evolution(ackley, bounds, recombination=recombination, seed=42))
data_recombination = pd.DataFrame(result_recombination).join(pd.DataFrame(recombinations))
data_recombination.columns=['x', 'fun', 'nfev', 'nit', 'message', 'success', 'recombination']
data_recombination
```

```python
plt.plot(data_recombination['recombination'], data_recombination['nit'])
```

```python
updatings = ['immediate', 'deferred']
result_updating = []
for updating in updatings:
    result_updating.append(differential_evolution(ackley, bounds, updating=updating, seed=42))
data_updating = pd.DataFrame(result_updating).join(pd.DataFrame(updatings))
data_updating.columns=['x', 'fun', 'nfev', 'nit', 'message', 'success', 'updating']
data_updating
```

```python
plt.barh(data_updating['updating'], data_updating['nit'])
```
