# Задание повышенного уровня.

Задан объект в пространстве 2D. 


```python
import numpy as np
import matplotlib.pyplot as plt

A = [
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
    ]
A = np.array(A)
x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Изменить положение объекта (сместить по оси x на 300 пикселей, по оси y на 200 пикселей).

```python
TransformArray = np.array([
    [1, 0, 300],
    [0, 1, 200],
    [0, 0, 1]
])

B = []
for point in A:
    new_point = np.dot(TransformArray, np.transpose(point))
    B.append(new_point)
B = np.array(B)
print(B)
```

```python
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([0, 500])
plt.xlim([100, 500])
```

Уменьшить объект в два раза.

```python
TransformArray = np.array([
    [0.5, 0, 0],
    [0, 0.5, 0],
    [0, 0, 1]
])

D = []
for point in A:
    new_point=np.dot(TransformArray,np.transpose(point))
    D.append(new_point)
D = np.array(D).astype(int)
print(D)
```

```python
x = D[:,0]
y = D[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Повернуть на 130 градусов.

```python
from math import cos, sin, pi

theta = 130 * 180/pi
TransformArray = np.array([
    [cos(theta), sin(theta), 0],
    [-sin(theta), cos(theta), 0],
    [0, 0, 1]
])

E = []
for point in A:
    new_point=np.dot(TransformArray,np.transpose(point))
    E.append(new_point)
E = np.array(E).astype(int)
print(E)
```

```python
x = E[:,0]
y = E[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Отразить объект относительно прямой y=x

```python
TransformArray = np.array([
    [0, 1, 0],
    [1, 0, 0],
    [0, 0, 1]
])

G = []
for point in A:
    new_point=np.dot(TransformArray,np.transpose(point))
    G.append(new_point)
G = np.array(G).astype(int)
print(G)
```

```python
x = G[:,0]
y = G[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1,2)

```python
TransformArray1 = np.array([
    [1, 0, 200],
    [0, 1, 300],
    [0, 0, 1]
])
TransformArray2 = np.array([
    [0.5, 0, 0],
    [0, 1.2, 0],
    [0, 0, 1]
])

F = []
for point in A:
    new_point = np.dot(TransformArray1, np.dot(TransformArray2, np.transpose(point)))
    F.append(new_point)
F = np.array(F).astype(int)
print(F)
```

```python
x = F[:,0]
y = F[:,1]
plt.plot(x, y)
plt.ylim([100, 500])
plt.xlim([100, 400])
```
