**Задание 1**
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.


```python
import pandas as pd
ratings = pd.read_csv('ratings.csv')
max_rating = ratings[ratings['rating'] == 5]['movieId'].value_counts().index[0]
max_rating
```

```python
movies = pd.read_csv('movies.csv')
movie = movies[movies['movieId'] == max_rating]
movie
```

```python
print(f'Фильму "', movie['title'].tolist()[0], '" было выставлено больше всего оценок 5.0')
```

**Задание 2**
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
power = pd.read_csv('power.csv')
```

```python
power[ power['country'].str.contains('L') ]['country'].unique()
```

```python
power[ power['country'].str.contains('Est') ]['country'].unique()
```

```python
selection = power[ (power['country'].isin(['Latvia', 'Lithuania', 'Estonia']))\
& (power['category'].isin([4, 12, 21])) & (power['year'] >= 2005) & (power['year'] <= 2010) & (power['quantity'] > 0)]
sum_power = selection['quantity'].sum()
print(f'Cуммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 \nза период с 2005 по 2010 год без учета отрицательных значений составило', sum_power)
```
